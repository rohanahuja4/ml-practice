# Reviewing moments of a distribution using Python (with ipynb)

In this notebook, we will review the concept of moments of a distribution and how to calculate them using Python. We will also see how to calculate the moments of a sample and how to estimate the moments of a population using a sample.

Let's first set up our imports and set up some data to work with:

```python
import numpy as np
from scipy.stats import skew, kurtosis
import pandas as pd
import matplotlib.pyplot as plt

# Generate random data
np.random.seed(0)
normal_data = np.random.normal(loc=0, scale=1, size=1000) # Standard normal
uniform_data = np.random.uniform(low=0, high=10, size=1000) # Uniform [0, 10]
skewed_data = np.random.exponential(scale=2, size=1000) # Exponential

# Display the distribution of the data
fig, ax = plt.subplots(1, 3, figsize=(15, 5))
ax[0].hist(normal_data, bins=50)
ax[0].set_title('Normal Distribution')
ax[1].hist(uniform_data, bins=50)
ax[1].set_title('Uniform Distribution')
ax[2].hist(skewed_data, bins=50)
ax[2].set_title('Exponential Distribution')
plt.show()
```

## Moments of a distribution

The $n$-th moment of a distribution is defined as:
$$
\mu_n = E[X^n] = \int_{-\infty}^{\infty} x^n f(x) dx
$$

where $f(x)$ is the probability density function of the distribution. The first moment is the mean, the second moment is the variance, and the third moment is the skewness. The fourth moment is called kurtosis.

## Moments of a sample

The $n$-th moment of a sample is defined as:
$$
\hat{\mu}_n = \frac{1}{N} \sum_{i=1}^{N} x_i^n
$$

where $x_i$ are the values of the sample and $N$ is the number of elements in the sample.

Let's calculate the moments of a sample using Python:

```python
# Calculate the skewness and kurtosis of a sample from each distribution
normal_skew = skew(normal_data)
normal_kurt = kurtosis(normal_data)
uniform_skew = skew(uniform_data)
uniform_kurt = kurtosis(uniform_data)
skewed_skew = skew(skewed_data)
skewed_kurt = kurtosis(skewed_data)
print(f'Normal Distribution: Skewness = {normal_skew:.2f}, Kurtosis = {normal_kurt:.2f}')
print(f'Uniform Distribution: Skewness = {uniform_skew:.2f}, Kurtosis = {uniform_kurt:.2f}')
print(f'Exponential Distribution: Skewness = {skewed_skew:.2f}, Kurtosis = {skewed_kurt:.2f}')
```
